```python
import json
import glob
import random
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
```

```python
def get_random_file(file_list):
    return file_list[random.randint(0, len(file_list) - 1)]
```

```python
file_list = glob.glob('./Data/data/*.json')
img_data = json.loads(open(get_random_file(file_list)).read())

normal_img = Image.open('./Data/dataset/' + img_data['file_name'])
```

```python
fig = plt.figure(figsize = (8, 8))

ax = fig.add_subplot(111)

for index, item in enumerate(img_data['arr_boxes']):
    box = patches.Rectangle((item['x'], item['y']),
                            item['width'], item['height'], 
                            linewidth = 2,
                            edgecolor = (random.uniform(0.0,1.0), random.uniform(0.0,1.0),random.uniform(0.0,1.0)),
                            facecolor = 'none', 
                            label = item['class'])
    ax.add_patch(box)
    ax.axis('off')

plt.legend()
plt.imshow(normal_img)
plt.show()
```
